```python
import tkinter as tk
from tkinter import ttk


class InputFrame(ttk.Frame):
    def __init__(self, container):
        super().__init__(container)
        # setup the grid layout manager
        self.columnconfigure(0, weight=1)
        self.columnconfigure(0, weight=3)
        self.__create_widgets()

    def __create_widgets(self):
        # High Res File
        ttk.Label(self, text='High Resolution Input').grid(column=0, row=0, sticky=tk.W)
        highres = ttk.Entry(self, width=24)
        highres.focus()
        highres.grid(column=1, row=0, sticky=tk.W)
        highres_find = ttk.Button(self, text='Find...')
        highres_find.focus()
        highres_find.grid(column=2, row=0)

        # Multispectral File
        ttk.Label(self, text='Multispectral Input').grid(column=0, row=1, sticky=tk.W)
        multi = ttk.Entry(self, width=24)
        multi.focus()
        multi.grid(column=1, row=1, sticky=tk.W)
        multi_find = ttk.Button(self, text='Find...')
        multi_find.focus()
        multi_find.grid(column=2, row=1)

        # Output File:
        ttk.Label(self, text='Output File').grid(column=0, row=9, sticky=tk.W)
        output = ttk.Entry(self, width=24)
        output.grid(column=1, row=9, sticky=tk.W)
        output_find = ttk.Button(self, text='Find...')
        output_find.focus()
        output_find.grid(column=2, row=9)

        # Generate Pyramids checkbox
        gen_pyramids = tk.StringVar()
        gen_pyramids_check = ttk.Checkbutton(
            self,
            text='Generate Pyramids',
            variable=gen_pyramids,
            command=lambda: print(gen_pyramids.get()))
        gen_pyramids_check.grid(column=0, row=3, sticky=tk.W)

        # Calculate Statistics checkbox
        calc_statistics = tk.StringVar()
        calc_statistics_check = ttk.Checkbutton(
            self,
            variable=calc_statistics,
            text='Calculate Statistics',
            command=lambda: print(calc_statistics.get()))
        calc_statistics_check.grid(column=0, row=4, sticky=tk.W)

        for widget in self.winfo_children():
            widget.grid(padx=0, pady=5)


class ButtonFrame(ttk.Frame):
    def __init__(self, container):
        super().__init__(container)
        # setup the grid layout manager
        self.columnconfigure(0, weight=1)

        self.__create_widgets()



    def __create_widgets(self):
        ttk.Button(self, text='Resample Method').grid(column=2, row=7)
        ttk.Button(self, text='Select Layer').grid(column=2, row=8)
        ttk.Button(self, text='Run').grid(column=2, row=9)
    

        for widget in self.winfo_children():
           widget.grid(padx=0, pady=3)


class App(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title('Pansharpening Tool')
        self.geometry('500x200')
        self.resizable(0, 0)

        # layout on the root window
        self.columnconfigure(0, weight=4)
        self.columnconfigure(1, weight=1)

        self.__create_widgets()

    def __create_widgets(self):
        
        # create the input frame
        input_frame = InputFrame(self)
        input_frame.grid(column=0, row=0)

        # create the button frame
        button_frame = ButtonFrame(self)
        button_frame.grid(column=1, row=0)


if __name__ == "__main__":
    app = App()
    app.mainloop()
```
